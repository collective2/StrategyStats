```r
stats = getC2ScoreDbData("select Return030 as Value from ReturnsDataInIntervalsCleaned 
    where Equity030 is not null and Age > 180 and Return180 > 0.1 and Return180 < 0.2")
cat("Number of observations =",NROW(stats)," SD = ",sd(stats$Value),"\n\n")
summary(stats$Value)
hist(stats$Value,breaks=50)
```

```r
data = stats  
n = nrow(data)
  mean = mean(data$Value) 
  sd = sd(data$Value)
  
  binwidth = 0.025 # passed to geom_histogram and stat_function
  
  ggplot(data, aes(x = Value)) +
    geom_histogram(binwidth = binwidth, colour = "white", fill = "cornflowerblue", size = 0.1) +
    stat_function(fun = function(x) dnorm(x, mean = mean, sd = sd) * n * binwidth, color = "darkred", size = 1)    
```
